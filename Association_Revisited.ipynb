# Step 1 Import Libraries
# decd.co/arules-apriori    for data

```python
import pandas as pd
import matplotlib.pyplot  as plt

# apriori function for frequest itemset and extarct possible rules
from mlxtend.frequent_patterns import apriori , association_rules

```

```python
data = pd.read_csv("Association Rules - Destination Recommendations - Destinations.csv")
```

```python
data.head()
```

# Step 3: Model Apriori

### find all frequent item sets using apriori algorith with a given support threshold

```python
frequent_itemsets=apriori(data, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets.info()
```

```python
frequent_itemsets.sort_values("support" , ascending=False )
```

### Now find association rules that can be constructed out of the frequest itemset using a given confidence threshold

```python
rules = association_rules(frequent_itemsets, min_threshold=0.01, metric='confidence')
```

```python
rules.sort_values("lift", ascending=False)
```

```python
# create new coloumn to tell us leght of antecedent and consquent

rules["antecedent_len"]=rules["antecedents"].apply(lambda x:len(x))
rules["consequent_len"]=rules["consequents"].apply(lambda x:len(x))
```

```python
rules.sort_values("lift", ascending=False)
```

```python

# fetch with 1 on LHS and 2 on RHS sorted by lift descenfding
rules[(rules['antecedent_len']==1 ) & (rules['consequent_len'] ==2) ].sort_values("lift", ascending = False)
```

```python
rules.info()
```

```python
rules[rules["antecedents"]=={"Dublin"} ].sort_values("lift", ascending = False)
```
