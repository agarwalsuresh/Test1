# 7. import functions from another library

```python
# get bokeh  function tomplot a figure and t read data arranged in columns

from bokeh.plotting import figure, ColumnDataSource

# get bokhen fucntions for outpouts that we want
from bokeh.io import output_file, show, output_notebook

# get bokeh fucntion to use a hover tool give information about data point

from bokeh.models import HoverTool

# Display bokeh figures in line inside Jupyter notebook
output_notebook()
```

```python
# Load hthe dataframe data to be use in boken 
source = ColumnDataSource(data)
```

Create a bokeh figure with a hover tool 

```python
p=figure(tools=['hover'])
```

Create the figure p by displaying data points as dots (i.e circles)

```python
p.circle('market_value', 'page_views', source=source )

# Specify the  data in the hover tool
hover=HoverTool(tooltips=[('Name', '@name'), ('Team', '@club')])
```

```python
p.add_tools(hover)
show(p)
```
